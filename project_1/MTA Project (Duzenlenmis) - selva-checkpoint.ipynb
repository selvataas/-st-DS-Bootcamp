```python
# Import Libraries

import numpy as np
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt , seaborn as sns
```

Newyork metro turnike datası http://web.mta.info/developers/turnstile.html adresinden çekilmek isteniyor.Haftalık verilerin her cumartesi olarak eklendiği görüldüğü için 2021 yılı mayıs aynın ilk iki haftası cumartesilerine ait days değişkeni oluşturuldu.

```python
days2018 = pd.date_range(start= '5/07/2018', end ='5/20/2018', freq ='W-SAT').tolist()   
days2019 = pd.date_range(start= '5/07/2019', end ='5/20/2019', freq ='W-SAT').tolist()
days2020 = pd.date_range(start= '5/07/2020', end ='5/20/2020', freq ='W-SAT').tolist()    
days2021 = pd.date_range(start= '5/07/2021', end ='5/20/2021', freq ='W-SAT').tolist() 
days = days2018 +  days2019  + days2020 + days2021                    

print(days)
```

days değişkenine göre listelenen tarihlere ait tüm url'ler listeye atılarak dataframe oluşturuldu.URL sayısı kontrol edildi.
Örnek url  http://web.mta.info/developers/data/nyct/turnstile/turnstile_211225.txt

```python
url_list = []
for date in days:    
    date_format = dt.strftime(date,'%y%m%d')  
    #print(date_format)
    url= 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_'+date_format+'.txt'         
    url_list.append(url)
    
df =pd.DataFrame(url_list)
print (url_list) 
print(len(url_list))
```

İstenen data url listesi üzerinden okunarak birleştirildi.

```python
data = pd.concat(map (pd.read_csv, url_list))
```

Tarih ve saat değişkenleri birleştirilerek yeni bir kolon eklendi.
Turnike bazında inceleme için TURNSTILE adlı kolon eklendi.

```python
data["DATETIME"] = pd.to_datetime(data["DATE"] +" "+data["TIME"])
data["TURNSTILE"] = data["C/A"] +""+data["UNIT"]+""+data["SCP"]#,+""+data["STATION"],
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

Veriler turnike ve tarih bazında sıralandı.

```python
data=data.sort_values(["TURNSTILE","DATETIME"],ascending = [True, True])
```

```python
data.head()
```

Kolon adlarında yer alan boşluklar kaldırıldı.

```python
data.columns=data.columns.str.strip()
```

Verilerin turnike ve gün bazında gruplanarak net giriş ve çıkışın bulunması gerekiyor.

Turnike net giriş ve net çıkış sayıları bulunarak 2 yeni kolon oluşturuldu.

```python
data["NET_ENTRIES"] = data["ENTRIES"].diff().fillna(data["ENTRIES"].iloc[0])
data["NET_EXITS"] = data["EXITS"].diff().fillna(data["EXITS"].iloc[0])
```

```python
pd.set_option('display.max_rows', None)
data.head(500)
```

Turnike sayısı bulundu.

Net giriş ve çıkışları çok fazla ve negatif çıkan sayılar bulunarak kendinden bir sonraki giriş veya çıkış sayısı olarak güncellendi.

```python
data[data['NET_ENTRIES']>1000]['NET_ENTRIES'].count() + data[data['NET_ENTRIES']< 0]['NET_ENTRIES'].count()
```

```python
data[data['NET_EXITS']>1000]['NET_EXITS'].count() + data[data['NET_EXITS']< 0]['NET_EXITS'].count()
```

```python
ENTRY_LIST =  list(data['NET_ENTRIES']) 
 
index = 0
for x  in ENTRY_LIST:
    if  x > 1000 or x < 0 :
        ENTRY_LIST[index]=ENTRY_LIST[index+1]      
    index = index + 1
    
data['CAL_NET_ENTRIES']  = ENTRY_LIST
```

```python
EXIT_LIST =  list(data['NET_EXITS']) 
index = 0
for  y  in EXIT_LIST:
    if  y > 1000 or y < 0 :
        EXIT_LIST[index]=EXIT_LIST[index+1]    
    index = index + 1
    
data['CAL_NET_EXITS']  = EXIT_LIST
    
```

```python
data["SUM_ENTRY_EXIT"] = data["CAL_NET_ENTRIES"] +  data["CAL_NET_EXITS"]
```
